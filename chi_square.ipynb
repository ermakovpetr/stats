# Распределение Хи-квадрат Пирсона для монетки

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'

import random
import scipy.stats as ss
import matplotlib.pyplot as plt

trials = 10000
N = 60

result = []

for _ in xrange(trials):
    experiment = [random.randint(0,1) for _ in xrange(N)]

    result.append(ss.chisquare(
            [sum(experiment), N - sum(experiment)], [N / 2., N / 2.]
        ).statistic)

plt.hist(result, bins=30)
plt.show()
```
